```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv('../data/dengue_csv/combined_dengue_data.csv')
df2=pd.read_csv('../data/dengue_csv/extract_dengue_data.csv',index_col=0)
df3=pd.read_csv('../data/dengue_csv/dengue_dataset.csv')
df4=pd.read_csv('../data/dengue_csv/Dengue-worldwide-dataset-modified.csv')
df5=pd.read_csv('../data/dengue_csv/dengue.csv')
```

```python
pre_df3=df3.drop(['Platelet_Count','White_Blood_Cell_Count','Gender'],axis=1)
pre_df3=pre_df3.rename(columns={'Temperature':'temp','Body_Pain':'muscle pain','Rash':'rash','Infected':'dengue'})
pre_df3 = pre_df3.replace({True: 1, False: 0})
pre_df3=pre_df3.astype(float)
pre_df3.head()
```

```python
df4.head()
```

```python
pre_df4=df4.drop(['dengue.days','dengue.p_i_d','dengue.date_of_fever','dengue.dwelling place ','dengue.wbc','dengue.hemoglobin','dengue._hematocri','dengue.platelet'],axis=1)
pre_df4=pre_df4.rename(columns={'current_temp':'temp',
                                'dengue.servere_headche':'headache',
                                'dengue.pain_behind_the_eyes':'eye pain',
                                'dengue.joint_muscle_aches':'muscle pain',
                                'dengue.metallic_taste_in_the_mouth':'metallic taste',
                                'dengue.appetite_loss':'appetite_loss',
                                'dengue.addominal_pain':'addminal pain',
                                'dengue.nausea_vomiting':'nausea_vomiting',
                                'dengue.diarrhoea':'diarrhoea',
                                'dengue_or_not':'dengue'})

pre_df4 = pre_df4.replace({'yes': 1, 'no': 0})
pre_df4=pre_df4.dropna()
pre_df4=pre_df4.astype(float)
pre_df4.head()
```
