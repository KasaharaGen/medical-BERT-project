```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 混同行列（％）
cm = np.array([
    [38.7, 61.2],  # TN, FP
    [10.3, 89.6]   # FN, TP
])

# ラベル
labels = [["38.7%", "61.2%"],
          ["10.3%", "89.6%"]]

# プロット
plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=labels,
    fmt="",
    cmap="Blues",
    cbar=True,
    annot_kws={"size": 16},   # 文字を大きく
    linewidths=1,
    linecolor="white"
)

plt.xlabel("Predicted Label", fontsize=14)
plt.ylabel("True Label", fontsize=14)
plt.title("Confusion Matrix ", fontsize=16)

plt.xticks([0.5, 1.5], ["0", "1"], fontsize=12)
plt.yticks([0.5, 1.5], ["0", "1"], fontsize=12, rotation=0)

plt.tight_layout()
plt.show()
```
